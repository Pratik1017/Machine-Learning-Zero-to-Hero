```python
import requests
```

```python
for i in range(1,51):
    a = requests.get(f"https://books.toscrape.com/catalogue/page-{i}.html")
    with open(f"htmls/page{i}.html" , "w" , encoding = "utf-8") as file:
        file.write(a.text)
        print(f"Downloaded Page {i} Succesfully")
```

```python
if a.status_code == 200:
    print("File Fetched Succesfully")
else:
    print("File Doesn't Fetched Succesfully")
```

```python
from bs4 import BeautifulSoup

with open("htmls/page1.html","r", encoding="utf-8") as f:
    content = f.read()
    
soup = BeautifulSoup(content, "html.parser")
```

```python
article = soup.find_all("article",class_="product_pod")
```

```python
items = []
for art in article:
    title = art.find("h3").find("a")["title"]
    price = art.select_one("p.price_color").text.split("Â£")[1]
    availability = art.select_one("p.instock.availability").text.strip()
    rating_star = art.select_one("p.star-rating")
    rating = rating_star["class"][1]
    items.append([title , price , availability , rating])
```

```python
items
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(items , columns=['Book_Title','Price','Availability', 'rating'])
```
