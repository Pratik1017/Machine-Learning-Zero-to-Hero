```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("=========== Zomato Sales Analysis========")
print("\nThe shape of Sales data: ",sales_data.shape)
print("\nThe first 3 resto data: \n",sales_data[:3])

# total sales per year

total_sales = np.sum(sales_data[:, 1:],axis=0)
print("\ntotal sales per year: ",total_sales)

# Minimum sales per restaurant
min_sales = np.min(sales_data[:,1:],axis = 1)
print("\nMinimum sales per restaurant: ",min_sales)

# Maximum sales per year
max_sales = np.max(sales_data[:,1:],axis = 1)
print("\nMaximum sales per restaurant: ",max_sales)

# Average sales per restautant
avg_sale = np.mean(sales_data[:,1:],axis = 1)
print("\nAverage sales per restaurant: ",avg_sale)
```

```python
# cumsum sales per restautant
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print("\ncumsum sales per restautant: \n",cumsum)

plt.figure(figsize = (10,6))
plt.plot(np.mean(cumsum[:,1:] , axis = 0))
plt.title("The Cumulative sales of all restro..")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector addition: ", vector1 + vector2)

print("\n Multiplication vector", vector1 * vector2)

print("\nDot Product", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print(angle)
```
