```python
import matplotlib.pyplot as plt

hours_studied = [2, 3, 5, 6, 7, 8, 10]
exam_score = [65, 70, 75, 80, 85, 90, 95]
color = ['red' for i in exam_score if i < 80]

fig, ax = plt.subplots()

# The 's' parameter controls marker size, 'alpha' controls transparency
ax.scatter(hours_studied, exam_score, s=100, alpha=0.7 , label = "hours_studied" , c =exam_score)

ax.set_title("Hours Studied vs. Exam Score")
ax.set_xlabel("Hours Studied")
ax.set_ylabel("Exam Score")
ax.legend(loc = 'upper left')
plt.show()
```

```python
Age = [1, 2, 3, 5, 6, 8, 10]
Price = [25, 22, 20, 18, 15, 12, 10]

fig, ax1 = plt.subplots()

# The 's' parameter controls marker size, 'alpha' controls transparency
ax1.scatter(Age, Price, s=100, alpha=0.7  , c =Price)
ax1.set_title("Car Price vs. Age")
ax1.set_xlabel("Price")
ax1.set_ylabel('Age')
ax1.legend(loc = 'upper left')
plt.show()
```
