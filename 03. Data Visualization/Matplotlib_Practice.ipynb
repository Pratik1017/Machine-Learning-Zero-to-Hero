```python
import matplotlib.pyplot as plt
```

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperature = [30, 32, 31, 33, 29, 28, 27]

clubs = ['Music', 'Drama', 'Coding', 'Sports']
students = [25, 18, 30, 22]

spending_categories = ['Rent', 'Food', 'Transport', 'Savings']
amounts = [800, 250, 100, 150]
explode = (0.1, 0, 0, 0) # Explode the 1st slice (Rent)


day_s = [1, 2, 3, 4, 5]
dev1_hours = [8, 6, 7, 8, 7]
dev2_hours = [0, 2, 1, 3, 4]
dev3_hours = [1, 1, 2, 0, 2]
# You must provide labels for the legend
labels = ["Developer 1", "Developer 2", "Developer 3"]


fig , axs = plt.subplots(3 , 1 , figsize = (6,9))
ax1 , ax2 , ax3 = axs.flatten()

ax1.plot(days , temperature , color = 'g' , marker = "o" , linestyle = '--')
ax1.set_title("Temp Basis on days")
ax1.set_xlabel("Days")
ax1.set_ylabel("temperature")
ax1.grid(True)


ax2.bar(clubs , students , color = 'g')
ax2.set_title("Student Enrollment in Clubs")
ax2.set_xlabel("Club")
ax2.set_ylabel("Number of Students")
ax2.grid(True)

ax3.pie(amounts , labels = spending_categories , explode = explode , autopct='%1.1f%%')
ax3.set_title("My Monthly Spending")
ax3.axis("equal")


plt.tight_layout()
plt.show()
```

```python
fig , ax4 = plt.subplots()
ax4.stackplot(day_s, dev1_hours, dev2_hours, dev3_hours, labels=labels)
ax4.set_title("Project Hours Contribution")
ax4.set_xlabel("Day")
ax4.set_ylabel("Cumulative Hours")
ax4.legend(loc='upper left') # Specify legend location
plt.show()
```

```python
Months = ['Jan', 'Feb', 'Mar', 'Apr']
Movies = [20, 25, 15, 30]
Books = [15, 10, 20, 15]
Games = [10, 15, 20, 25]
label1 = ["Movies" , "Books" , "Games"]

fig , ax5 = plt.subplots()
ax5.stackplot(Months , Movies , Books , Games , labels = label1)
ax5.set_title("Months Contribution")
ax5.set_xlabel("Months")
ax5.set_ylabel("Activity")
ax5.legend(loc='upper left') # Specify legend location
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 

# Generate 100 random scores between 0 and 100
scores = np.random.normal(70, 15, 100) # (mean, std_dev, count)

fig, ax6 = plt.subplots()

# The 'bins' parameter controls how many ranges to split the data into.
ax6.hist(scores, bins=10, edgecolor='black')

ax6.set_title("Distribution of Exam Scores")
ax6.set_xlabel("Score")
ax6.set_ylabel("Number of Students")
plt.show()
```
